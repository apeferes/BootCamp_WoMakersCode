### 1. Considere pessoas fluentes em inglês, qual a probabilidade complementar? Ou seja, qual a probabilidade de escolhermos uma pessoa aleatória e ela não ser fluente em inglês. Considere fluente quem tem o nível avançado.

```python
# Revendo as quantidades por nível
df['nível de proficiência em inglês'].value_counts()
```

```python
contagem_total = df['nível de proficiência em inglês'].count() # Contagem total
contagem_avancado = (df['nível de proficiência em inglês'] == 'Avançado').sum() # Contagem dos candidatos fluentes

porcentagem_avancado = contagem_avancado / contagem_total # Calulo de porcentagem
porcentagem_avancado # Probabilidade de ser fluente em inglês
```

### 3. Descubra a probabilidade de uma pessoa ter ensino superior completo no estado do Amazonas. Qual a probabilidade da quinta pessoa que você conversar, que é amazonense, ter ensino superior completo?

```python
df['escolaridade'].unique()
```

```python
# Quantas pessoas no amazonas tem no total ensino superior?

amazonas = df.loc[df['estado'].isin(['AM'])]
cont_amazonas = amazonas['estado'].count()

amazonas_ensino_superior = amazonas.loc[amazonas['escolaridade'].isin(['Superior', 'Pós-graduação'])]
cont_amazonas_ensino_superior = amazonas_ensino_superior['estado'].count()

# porcentagem de pessoas que tenham ensino superior no AM

am_ensino_superior_percent = cont_amazonas_ensino_superior / cont_amazonas
am_ensino_superior_percent
```

### 4. Considerando a renda das pessoas do nosso conjunto, podemos dizer que a renda de uma pessoa brasileira está na sua maioria em que faixa (faça faixa de 1.500 reais)? Qual é a sua função densidade de probabilidade?

```python
df.head
```

```python
# Descobrindo qual é a renda máxima para criação da clusterização

df['renda'].max()
```

```python
# Soma do total da renda
total_renda = len(df['renda'])

# Criando as variáveis de clusters
clusters = {
    1500: 0,
    3000: 0,
    4500: 0,
    6000: 0
}

# Criando as regras para a soma dos cidadãos por renda
for renda in df['renda']:
    for limite, cluster in clusters.items():
        if renda <= limite:
            clusters[limite] += 1
            break

# Calculando a probabilidade de densidade de renda para cada cluster
for limite, cluster in clusters.items():
    probabilidade = (cluster / total_renda) * 100
    print(f'A probabilidade da densidade de renda de {limite - 1500} até {limite} é de {probabilidade:.2f}%')
```

### 5. Calcule a média e a variância da renda da amostra. Depois faça a distribuição normal, inclua o gráfico.

```python
# Cálculo da média e da variância
media_renda = df['renda'].mean()
variancia_renda = df['renda'].var()

print(f"Média da renda: {media_renda:.2f}")
print(f"Variância da renda: {variancia_renda:.2f}")
```

```python
# Geração da distribuição normal
x = np.linspace(df['renda'].min(), df['renda'].max(), 100)
y = norm.pdf(x, media_renda, np.sqrt(variancia_renda))

# Plotando o gráfico da distribuição normal
plt.plot(x, y)
plt.title('Distribuição Normal da Renda')
plt.xlabel('Renda')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True)
plt.show()
```

### 6. Primeiro considere a probabilidade encontrada no nosso conjunto de pessoas com escolaridade de pós-graduação. Considerando a amostra de população brasileira com 1 milhão de habitantes, qual a probabilidade de encontrarmos 243 mil pessoas com pós-graduação?

```python
df.head
```

```python
df_escolaridade = df['escolaridade'].value_counts()
df_escolaridade
```

```python
total_escolaridade = df_escolaridade.sum()
porcentagem_por_tipo_escolaridade = (df_escolaridade / total_escolaridade)
print(porcentagem_por_tipo_escolaridade)
```

```python
from scipy.stats import binom

n = 1000000  # Número total de habitantes
k = 243000  # Número desejado de pessoas com pós-graduação
p = porcentagem_por_tipo_escolaridade['Pós-graduação']  # Probabilidade de uma pessoa ter pós-graduação

# Calculando a probabilidade
probabilidade = binom.pmf(k, n, p)

print(f"A probabilidade de encontrar exatamente {k} pessoas com pós-graduação em uma população de {n} habitantes é: {probabilidade}")
```

### 7. Somando as densidades nós temos a função de densidade acumulada. Considerando a coluna ‘Escolaridade’ faça a função de densidade acumulada discreta para cada nível de escolaridade.

```python
ordem_escolaridade = {
    'Fundamental': 1,
    'Médio': 2,
    'Superior': 3,
    'Pós-graduação': 4
}

# Mapeando os níveis de escolaridade para uma variável numérica que reflete a ordem educacional
df['escolaridade_Ord'] = df['escolaridade'].map(ordem_escolaridade)

# Ordenando o DataFrame pela nova coluna de ordem
df_ordenado = df.sort_values('escolaridade_Ord')

# Calculando a proporção acumulada
proporcao_acumulada_correta = df_ordenado.groupby('escolaridade_Ord')['escolaridade'].count().cumsum() / len(df_ordenado)

# Mapeando os números de volta para os níveis de escolaridade
proporcao_acumulada_correta.index = proporcao_acumulada_correta.index.map({v: k for k, v in ordem_escolaridade.items()})

print(proporcao_acumulada_correta)
```

### 8. Qual a margem de erro amostral da proporção populacional considerando a proporção de pessoas com nível de inglês intermediário?

```python
df.head(10)
```

```python
contagem_total = df['nível de proficiência em inglês'].count() # Contagem total
contagem_intermediario = (df['nível de proficiência em inglês'] == 'Intermediário').sum()

print(contagem_total,contagem_intermediario)
```

```python
# Dados da amostra
n = contagem_total  # Tamanho da amostra
p = contagem_intermediario / contagem_total  # Proporção de pessoas com inglês intermediário

# Nível de confiança de 95%, Z = 1.96 para distribuição normal padrão
Z = 1.96

# Calculando a margem de erro
ME = Z * np.sqrt((p * (1 - p)) / n)

print(f"A margem de erro amostral é: {ME:.4f}")
```

### 9. Calcula a renda da população. Qual a probabilidade de encontrar 60 pessoas com uma renda mil reais superior à média?

```python
from scipy.stats import norm, binom

media_renda = df['renda'].mean()  # média da renda
desvio_padrao_renda = df['renda'].std()  # desvio padrão da renda
renda_alvo = media_renda + 1000  # renda alvo, 1000 reais acima da média

# Calculando a probabilidade de uma pessoa ter renda > renda_alvo
p = 1 - norm.cdf(renda_alvo, media_renda, desvio_padrao_renda)

# Tamanho da amostra
n = df['renda'].count()

# Número desejado de pessoas com renda > renda_alvo
k = 60

# Calculando a probabilidade usando a distribuição binomial
probabilidade = binom.pmf(k, n, p)

print(f"Probabilidade de encontrar 60 pessoas com renda mil reais acima da média: {probabilidade:.4f}")
```
